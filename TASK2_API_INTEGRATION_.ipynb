```python
#  Cell 1 — Imports (requests is pre-installed in Colab)
# ============================================================
import requests, json
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets
```

```python
#  Cell 2 — Fetch data
# ============================================================
BASE_URL = "https://jsonplaceholder.typicode.com/posts"

def fetch_posts(url):
    response = requests.get(url, timeout=10)
    response.raise_for_status()
    data = response.json()
    print(f"✅ Fetched {len(data)} posts  |  Status: {response.status_code}")
    return data

all_posts = fetch_posts(BASE_URL)

print("\n── Sample record ──────────────────────────────────────")
print(json.dumps(all_posts[0], indent=2))
```

```python
#  Cell 3 — Display helper (HTML table)
# ============================================================
def show_table(posts, max_rows=10):
    if not posts:
        print("⚠️  No results found.")
        return

    subset = posts[:max_rows]

    header = """
    <style>
      .post-table { width:100%; border-collapse:collapse; font-family:Arial,sans-serif; font-size:13px; margin-top:8px; }
      .post-table th { background:#3B82F6; color:#fff; padding:8px 12px; text-align:left; border:1px solid #2563EB; }
      .post-table td { padding:7px 12px; border:1px solid #e2e8f0; vertical-align:top; }
      .post-table tr:nth-child(even) td { background:#F8FAFC; }
      .post-table tr:hover td { background:#EFF6FF; }
      .badge { display:inline-block; padding:2px 8px; border-radius:12px; background:#DBEAFE; color:#1D4ED8; font-size:11px; font-weight:600; }
    </style>
    """

    rows = ""
    for p in subset:
        rows += f"""
        <tr>
          <td style='width:40px;text-align:center'>{p['id']}</td>
          <td style='width:70px;text-align:center'><span class='badge'>User {p['userId']}</span></td>
          <td style='width:40%'><strong>{p['title']}</strong></td>
          <td>{p['body'][:100]}…</td>
        </tr>"""

    table = f"""
    {header}
    <p style='font-family:Arial;font-size:12px;color:#64748B;margin:4px 0'>
      Showing <b>{len(subset)}</b> of <b>{len(posts)}</b> posts
    </p>
    <table class='post-table'>
      <thead><tr><th>ID</th><th>User</th><th>Title</th><th>Body</th></tr></thead>
      <tbody>{rows}</tbody>
    </table>
    """
    display(HTML(table))


print("\n── All Posts (first 10) ────────────────────────────────")
show_table(all_posts, max_rows=10)
```

```python
#  Cell 4 — Search & Filter functions
# ============================================================
def search_keyword(posts, keyword):
    kw = keyword.strip().lower()
    if not kw:
        return posts
    return [p for p in posts if kw in p["title"].lower() or kw in p["body"].lower()]

def filter_user(posts, user_id):
    if user_id == 0:
        return posts
    return [p for p in posts if p["userId"] == user_id]

def filter_id_range(posts, start, end):
    return [p for p in posts if start <= p["id"] <= end]


print("\n── Keyword search: 'qui' ───────────────────────────────")
show_table(search_keyword(all_posts, "qui"), max_rows=5)

print("\n── Posts by User 3 ─────────────────────────────────────")
show_table(filter_user(all_posts, 3), max_rows=5)

print("\n── Posts with ID 11–20 ─────────────────────────────────")
show_table(filter_id_range(all_posts, 11, 20), max_rows=10)
```

```python
#  Cell 5 — Interactive widget
# ============================================================
keyword_box = widgets.Text(
    value="", placeholder="Type keyword…",
    description="🔍 Search:", layout=widgets.Layout(width="320px")
)
user_drop = widgets.Dropdown(
    options=[("All users", 0)] + [(f"User {i}", i) for i in range(1, 11)],
    value=0, description="👤 User:", layout=widgets.Layout(width="200px")
)
rows_slider = widgets.IntSlider(
    value=10, min=1, max=50, step=1,
    description="Rows:", layout=widgets.Layout(width="340px")
)
out = widgets.Output()

def refresh(_=None):
    filtered = search_keyword(all_posts, keyword_box.value)
    filtered = filter_user(filtered, user_drop.value)
    with out:
        clear_output(wait=True)
        show_table(filtered, max_rows=rows_slider.value)

keyword_box.observe(refresh, names="value")
user_drop.observe(refresh, names="value")
rows_slider.observe(refresh, names="value")

print("\n── Interactive Filter ───────────────────────────────────")
display(widgets.VBox([widgets.HBox([keyword_box, user_drop]), rows_slider, out]))
refresh()

```

```python
  #Cell 6 — Export filtered results to JSON
# ============================================================
def export_json(posts, filename="filtered_posts.json"):
    with open(filename, "w") as f:
        json.dump(posts, f, indent=2)
    print(f"💾 Saved {len(posts)} records → '{filename}'")
    try:
        from google.colab import files
        files.download(filename)
        print("⬇️  Download triggered.")
    except ImportError:
        print("   (File saved locally — not in Colab)")

export_json(search_keyword(all_posts, "qui"), "filtered_posts.json")
```
