```python
import os
import shutil
import logging
from datetime import datetime

# Setup logging
logging.basicConfig(filename="automation.log", level=logging.INFO,
                    format="%(asctime)s - %(message)s")

# File type categories
CATEGORIES = {
    "Images": [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".svg"],
    "Documents": [".pdf", ".doc", ".docx", ".txt", ".xlsx", ".csv"],
    "Videos": [".mp4", ".mkv", ".avi", ".mov"],
    "Audio": [".mp3", ".wav", ".aac"],
    "Archives": [".zip", ".tar", ".gz", ".rar"],
    "Code": [".py", ".js", ".html", ".css", ".json"],
}

def get_category(ext):
    for category, extensions in CATEGORIES.items():
        if ext.lower() in extensions:
            return category
    return "Others"

def sort_files(folder):
    """Sort files into subfolders by type."""
    moved = 0
    for filename in os.listdir(folder):
        filepath = os.path.join(folder, filename)
        if not os.path.isfile(filepath):
            continue
        ext = os.path.splitext(filename)[1]
        category = get_category(ext)
        dest_dir = os.path.join(folder, category)
        os.makedirs(dest_dir, exist_ok=True)
        shutil.move(filepath, os.path.join(dest_dir, filename))
        logging.info(f"Moved: {filename} → {category}/")
        moved += 1
    print(f"✅ Sorted {moved} files.")

def rename_files(folder, prefix):
    """Rename files with a prefix + timestamp."""
    renamed = 0
    for i, filename in enumerate(os.listdir(folder)):
        filepath = os.path.join(folder, filename)
        if not os.path.isfile(filepath):
            continue
        ext = os.path.splitext(filename)[1]
        new_name = f"{prefix}_{i+1}{ext}"
        os.rename(filepath, os.path.join(folder, new_name))
        logging.info(f"Renamed: {filename} → {new_name}")
        renamed += 1
    print(f"✅ Renamed {renamed} files.")

def delete_empty_folders(folder):
    """Remove empty subdirectories."""
    removed = 0
    for dirpath, dirnames, filenames in os.walk(folder, topdown=False):
        if dirpath == folder:
            continue
        if not os.listdir(dirpath):
            os.rmdir(dirpath)
            logging.info(f"Deleted empty folder: {dirpath}")
            removed += 1
    print(f"✅ Removed {removed} empty folders.")

def show_log():
    """Display the last 10 log entries."""
    try:
        with open("automation.log") as f:
            lines = f.readlines()
        print("\n📋 Last 10 log entries:")
        print("".join(lines[-10:]))
    except FileNotFoundError:
        print("No log file found yet.")

# ── Main Menu ──────────────────────────────────────────────
def main():
    print("\n🗂️  File Automation Script")
    folder = input("Enter folder path (or press Enter for current dir): ").strip() or "."

    if not os.path.isdir(folder):
        print("❌ Invalid folder path.")
        return

    print("\nOptions:\n 1. Sort files by type\n 2. Rename files\n 3. Delete empty folders\n 4. Show logs\n 5. Exit")
    choice = input("Choose (1-5): ").strip()

    if choice == "1":
        sort_files(folder)
    elif choice == "2":
        prefix = input("Enter prefix for renamed files: ").strip() or "file"
        rename_files(folder, prefix)
    elif choice == "3":
        delete_empty_folders(folder)
    elif choice == "4":
        show_log()
    elif choice == "5":
        print("Bye!")
    else:
        print("❌ Invalid choice.")

if __name__ == "__main__":
    main()
```
